```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import regex as re
```

```python
geojson = gpd.read_file('./input/ghmc-data/GHMC.geojson')
geojson = geojson[['id', 'name', 'geometry']]
geojson.head()
```

let us use ward numbers as identifiers as they are unique

```python
geojson['name'] = geojson['name'].str.lstrip("Ward")
geojson['ward_no'] = geojson.name.str.extract(r'(\d+)')
geojson['ward_name'] = geojson.name.str.split(" ").str[2:].str.join(" ")
geojson['ward_name'] = geojson['ward_name'].str.upper()
geojson.head()
```

```python
ghmc_data = pd.read_csv("./input/ghmc-data/geolocations_amenities_ghmc.csv")
ghmc_data.dropna(inplace=True)
ghmc_data.head()
```

```python
ghmc_data['Wardno Name'] = ghmc_data['Wardno Name'].str.replace("-", " ")
ghmc_data['ward_no'] = ghmc_data['Wardno Name'].str.extract(r'(\d+)')
ghmc_data['ward_name'] = ghmc_data['Wardno Name'].str.split(" ").str[1:].str.join(" ")
```

```python
ghmc_data.head()
```

```python
categories = ghmc_data.Category.unique()
best_and_worst = pd.DataFrame(columns=['category','best_performer','worst_performer'])
for category in categories:
    temp_df = ghmc_data.loc[ghmc_data.Category == category]
    best_and_worst_numbers = temp_df.ward_name.value_counts()
    new_row = {'category': category,'best_performer': {best_and_worst_numbers.axes[0][0]:best_and_worst_numbers[0]}, 'worst_performer': {best_and_worst_numbers.axes[0][-1]:best_and_worst_numbers[-1]}}
    best_and_worst = best_and_worst.append(new_row, ignore_index=True)
    print(new_row)
```
